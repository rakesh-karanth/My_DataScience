```python
df = pd.read_csv("mercedes.csv", usecols=['X0', 'X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'X8'])
```

```python
for item in df.columns:
    print(item ,':', len(df[item].unique()))
```

Except X3, X4 remaining features contains mulpiple values. Using get dummies is not an option as it will result in creating more features.

###### encoding X3, X4 using get_dummies method.

Syntax:

pd.get_dummies(data_set, columns=['col1', 'col2'], drop_first= True)

'columns' attribute is used when particular column needs to be encoded. If not used all the categorical column will be encoded.

```python
pd.get_dummies(df, columns=['X3', 'X4'], drop_first=True)
```

Encoding 'X0', 'X1', 'X2','X5', 'X6', 'X8' columns. These columns contains multiple variables get_dummies method will create multiple columns, which will reduce the performance of the model.

One way is to select the top n most frequently used labels from the column and encode it. Remaining labels will be treated as zeroes.

```python
df.X0.value_counts().sort_values(ascending = False)
```

To apply it on all the categorical features present in the data set.

```python
top_10_X0 = [x for x in df.X0.value_counts().sort_values(ascending = False).head(10).index]
top_10_X0
```

```python
def onehot_encode_top_x(df, col, top_x_label):
    for label in top_x_label:
        df[col+'_'+label] = np.where(df[col]==label, 1,0)
        
```

```python
onehot_encode_top_x(df, 'X0', top_10_X0)
```

```python
df.head(25)
```

```python
top_10_X1 = [x for x in df.X1.value_counts().sort_values(ascending = False).head(10).index]
top_10_X1
```

```python
onehot_encode_top_x(df, 'X1', top_10_X1)
```
