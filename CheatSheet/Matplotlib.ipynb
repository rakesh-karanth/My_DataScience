```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,46,46,49]
min_t =[43,42,40,44,33,35,37]
avg = [45,48,48,46,40,42,41]
```

```python
plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg)

plt.xlabel('Week Days')
plt.ylabel('Temperature')
plt.title('Day vs Temp')
```

```python
plt.plot(days,max_t, label = 'max_t')
plt.plot(days,min_t, label = 'min_t')
plt.plot(days,avg, label='avg_t')

plt.xlabel('Week Days')
plt.ylabel('Temperature')
plt.title('Day vs Temp')
plt.legend()
```

```python
plt.plot(days,max_t, label = 'max_t')
plt.plot(days,min_t, label = 'min_t')
plt.plot(days,avg, label='avg_t')

plt.xlabel('Week Days')
plt.ylabel('Temperature')
plt.title('Day vs Temp')
plt.legend(loc = 'upper right', shadow = True)
```

### Bar graph

```python
company = ['GOOGL', 'AMZN', 'MSFT', 'FB']
revenue = [90,136,89,27]
```

```python
plt.bar(company, revenue)
```

```python
plt.bar(company,revenue, label = 'revenue', color = 'r', width = 0.1)
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("Company revenue")
plt.legend()
plt.grid(alpha=0.1)
```

```python
company = ['GOOGL', 'AMZN', 'MSFT', 'FB']
revenue = [90,136,89,27]
expenditure =[80,120,70,44]
```

```python
plt.bar(company,revenue, label = 'revenue', color = 'r', width = 0.1)
plt.bar(company,expenditure, label = 'expenditure', width = 0.1)
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("Company revenue")
plt.legend()
plt.grid(alpha=0.3)
```

```python
import numpy as np
xpos = np.arange(len(company))
```

```python
plt.bar(xpos-0.2, revenue, label = 'revenue', color = 'r', width = 0.1)
plt.bar(xpos+0.2, expenditure, label = 'expenditure', width = 0.1)
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("Company revenue")
plt.legend()
plt.grid(alpha=0.3)
```

#### To print company name use "xticks instead of xlabel"

```python
plt.bar(xpos-0.2, revenue, label = 'revenue', color = 'r', width = 0.3)
plt.bar(xpos+0.2, expenditure, label = 'expenditure', width = 0.3)

plt.xticks(xpos, company)

plt.ylabel("Revenue")
plt.title("Company revenue")
plt.legend()
plt.grid(alpha=0.3)
```

#### Horizontal plot

```python
plt.barh(company, revenue)
```

```python
plt.barh(xpos-0.2, revenue, label = 'revenue', color = 'r')
plt.barh(xpos+0.2, expenditure, label = 'expenditure')

plt.yticks(xpos, company)

plt.xlabel("Revenue")
plt.title("Company revenue")
plt.legend()
plt.grid(alpha=0.3)
```

### Histogram
- X axis is a variable and y axis is the frequency

```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
# - how many are normal [80-100]
# - how many are prediabetic [100-120]
# - how many are diabetic [125+]
```

```python
plt.hist(blood_sugar)
```

```python
plt.hist(blood_sugar, bins=20) # bins by default is 10
```

```python
plt.hist(blood_sugar, bins = [80,100,125,150])
```

- how many are normal [80-100] - 6
- how many are prediabetic [100-120] - 2
- how many are diabetic [125+] - 4

```python
plt.hist(blood_sugar, bins = [80,100,125,150], rwidth = 0.95)
```

```python
plt.hist(blood_sugar, bins = [80,100], rwidth = 0.70)
plt.hist(blood_sugar, bins = [100,125], rwidth = 0.70)
plt.hist(blood_sugar, bins = [125,150], rwidth = 0.70)
```

### Horizontal histogram

```python
plt.hist(blood_sugar, bins = 15, orientation='horizontal')
```

```python
plt.hist(blood_sugar, bins = [80,100,125,150], rwidth = 0.95, orientation='horizontal')
```

### Subplot

```python
year=[2014,2015,2016,2017]
income = [4000,4500, 5300,4600]
expense = [2800,3000,2800,3400]
```

```python
plt.subplot(1,2,1)   #plt.subplot(no. of row, no.of col, position)
plt.bar(year, expense,color='green')
plt.subplot(1,2,2)
plt.bar(year, income,color='y')
```

```python
plt.subplot(2,1,2)   #plt.subplot(no. of row, no.of col, position)
plt.bar(year, expense,color='green')
plt.subplot(2,2,1)
plt.bar(year, income,color='y')
```

```python
plt.subplot(1,2,1)   #plt.subplot(no. of row, no.of col, position)
plt.bar(year, expense,color='green')
plt.subplot(1,2,2)
plt.scatter(year, expense,color='y')
```
