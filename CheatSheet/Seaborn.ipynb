```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

### Relplot (relation plot)

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips)
```

Relplot can find the relation between max of 5 features

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'day')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'day', style= 'time')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'day', style= 'time', size = 'sex')
```

```python
iris = sns.load_dataset('iris')
```

```python
sns.scatterplot(x=iris.sepal_length, y= iris.petal_length)
```

### Bar plot

```python
titanic = sns.load_dataset('titanic')
```

```python
sns.barplot('class', 'age', data = titanic)
```

### Catplot (Categorical plot)

```python
sns.catplot('day', 'total_bill',data = tips)
```

```python
sns.catplot('day', 'total_bill',data = tips, jitter = False)
```

Below plot can be used to detect outlier

```python
sns.catplot('day', 'total_bill',data = tips, kind= 'box')
```

```python
sns.catplot('day', 'total_bill',data = tips, kind= 'swarm')
```

```python
sns.catplot('day', 'total_bill',data = tips, kind= 'violin')
```

### Box plot

```python
sns.boxplot('day', 'total_bill',data = tips)
```

### Distribution plot

```python
import numpy as np
x = np.random.randn(100)
x
```

```python
sns.displot(x)
```

```python
sns.displot(x, kde= True)
```

```python
sns.displot(x, kind = 'kde')
```

### Regplot - regression plot

```python
sns.regplot(x = 'total_bill', y = 'tip', data = tips)
```

### Pair plot

```python
sns.pairplot(tips)
```
