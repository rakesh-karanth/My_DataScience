```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,46,46,49]
min_t =[43,42,40,44,33,35,37]
avg = [45,48,48,46,40,42,41]
```

```python
plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg)

plt.xlabel('Week Days')
plt.ylabel('Temperature')
plt.title('Day vs Temp')
```

```python
plt.plot(days,max_t, label = 'max_t')
plt.plot(days,min_t, label = 'min_t')
plt.plot(days,avg, label='avg_t')

plt.xlabel('Week Days')
plt.ylabel('Temperature')
plt.title('Day vs Temp')
plt.legend()
```

```python
plt.plot(days,max_t, label = 'max_t')
plt.plot(days,min_t, label = 'min_t')
plt.plot(days,avg, label='avg_t')

plt.xlabel('Week Days')
plt.ylabel('Temperature')
plt.title('Day vs Temp')
plt.legend(loc = 'upper right', shadow = True)
```

```python
#grid

plt.plot(days,max_t, label = 'max_t')
plt.plot(days,min_t, label = 'min_t')
plt.plot(days,avg, label='avg_t')

plt.xlabel('Week Days')
plt.ylabel('Temperature')
plt.title('Day vs Temp')
plt.legend(shadow = True)
plt.grid(alpha = 0.4)
```

### Bar graph

```python
company = ['GOOGL', 'AMZN', 'MSFT', 'FB']
revenue = [90,136,89,27]
```

```python
plt.bar(company, revenue)
```

```python
plt.bar(company,revenue, label = 'revenue', color = 'r', width = 0.1)
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("Company revenue")
plt.legend()
plt.grid(alpha=0.1)
```

```python
company = ['GOOGL', 'AMZN', 'MSFT', 'FB']
revenue = [90,136,89,27]
expenditure =[80,120,70,44]
```

```python
plt.bar(company,revenue, label = 'revenue', color = 'r', width = 0.1)
plt.bar(company,expenditure, label = 'expenditure', width = 0.1)
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("Company revenue")
plt.legend()
plt.grid(alpha=0.3)
```

```python
import numpy as np
xpos = np.arange(len(company))
```

```python
plt.bar(xpos-0.2, revenue, label = 'revenue', color = 'r', width = 0.1)
plt.bar(xpos+0.2, expenditure, label = 'expenditure', width = 0.1)
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("Company revenue")
plt.legend()
plt.grid(alpha=0.3)
```

#### To print company name use "xticks instead of xlabel"

```python
plt.bar(xpos-0.2, revenue, label = 'revenue', color = 'r', width = 0.3)
plt.bar(xpos+0.2, expenditure, label = 'expenditure', width = 0.3)

plt.xticks(xpos, company)

plt.ylabel("Revenue")
plt.title("Company revenue")
plt.legend()
plt.grid(alpha=0.3)
```

#### Horizontal plot

```python
plt.barh(company, revenue)
```

```python
plt.barh(xpos-0.2, revenue, label = 'revenue', color = 'r')
plt.barh(xpos+0.2, expenditure, label = 'expenditure')

plt.yticks(xpos, company)

plt.xlabel("Revenue")
plt.title("Company revenue")
plt.legend()
plt.grid(alpha=0.3)
```
