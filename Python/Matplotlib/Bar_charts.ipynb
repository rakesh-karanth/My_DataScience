```python
import matplotlib.pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

```python
ages_x = [22, 23, 24, 25, 26, 27, 28, 29, 30]
```

```python
py_dev_y = [20046, 17100, 20000, 24744, 30500, 37732, 41247, 45372, 48876]
```

```python
plt.bar(ages_x, py_dev_y, label = 'python salary')
plt.title('Salary in USD vs age')
plt.xlabel('Age')
plt.ylabel('Salary in USD')
plt.legend()
plt.show()
```

# Use numpy to make plot look clear

```python
import numpy as np
```

```python
ages_x = [22, 23, 24, 25, 26, 27, 28, 29, 30]
```

```python
x_indexes = np.arange(len(ages_x))
```

```python
width = 0.25
```

#### Same as above plot but xlable is modified.  Instead of giving 'ages_x' as xlabel I used 'numpy length' as xlabel.  x_indexes-width and x_indexes+width gives an offset.

```python
py_dev_y = [20046, 17100, 20000, 24744, 30500, 37732, 41247, 45372, 48876]
plt.bar(x_indexes-width, py_dev_y, width = width, label = 'python salary')

js_dev_y = [16446, 16791, 18942, 21780, 25704, 29000, 34372, 37810, 43515]
plt.bar(x_indexes, js_dev_y, width = width, label = 'Javascript Salary') 

dev_y = [17784, 16500, 18012, 20628, 25206, 30252, 34368, 38496, 42000]
plt.bar(x_indexes+width, dev_y, width = width, label = 'developer Salary')

plt.title('Salary in USD vs age')
plt.xlabel('Age')
plt.ylabel('Salary in USD')
plt.legend()
plt.show()
```
