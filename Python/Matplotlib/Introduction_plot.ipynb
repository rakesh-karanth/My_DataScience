```python
import matplotlib.pyplot as plt
```

```python
ages_x = [22, 23, 24, 25, 26, 27, 28, 29, 30]
```

```python
py_dev_y = [20046, 17100, 20000, 24744, 30500, 37732, 41247, 45372, 48876]
```

# Plot Syntax

plt.plot(x_label, y_label, color = '< >', marker = '< >'
         
         linestyle = '< >', linewidth ='<number>', 
         
         label = '<give name for the particulat plot>')

# Simple plot for one vs one

```python
plt.plot(ages_x, py_dev_y)
plt.title('Salary in USD / age')
plt.xlabel('Age')
plt.ylabel('Salary in USD')
plt.show()
```

# plot for one x label vs multiple y label

```python
js_dev_y = [16446, 16791, 18942, 21780, 25704, 29000, 34372, 37810, 43515]
plt.plot(ages_x, js_dev_y, label = 'Javascript Salary') 

dev_y = [17784, 16500, 18012, 20628, 25206, 30252, 34368, 38496, 42000]
plt.plot(ages_x, dev_y, label = 'developer Salary')

"""rather than passing arguments to plt.legend use "label" 
so that it will not mess up the orders."""

plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by Age')
plt.legend()
plt.show()
```

# Formatting the plot lines

```python
js_dev_y = [16446, 16791, 18942, 21780, 25704, 29000, 34372, 37810, 43515]
plt.plot(ages_x, js_dev_y, label = 'Javascript Salary', color = 'k', linestyle = '--', marker = '.', linewidth = '2') 

dev_y = [17784, 16500, 18012, 20628, 25206, 30252, 34368, 38496, 42000]
plt.plot(ages_x, dev_y, label = 'developer Salary', color = 'g', marker = 'o', linewidth = '3')

"""rather than passing arguments to plt.legend use "label" 
so that it will not mess up the orders."""

plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by Age')
plt.legend() 
plt.show()
```

#### Legend location changing
- plt.legend(loc ='upper left')
- 'lowr left', 'upper right', 'upper left'
- plt.legend(loc=(0.07, 0..5))

# Filling area between the line plot

```python
js_dev_y = [16446, 16791, 18942, 21780, 25704, 29000, 34372, 37810, 43515]
plt.plot(ages_x, js_dev_y, label = 'Javascript Salary', color = 'k', linestyle = '--', marker = '.', linewidth = '2') 

dev_y = [17784, 16500, 18012, 20628, 25206, 30252, 34368, 38496, 42000]
plt.plot(ages_x, dev_y, label = 'developer Salary', color = 'g', marker = 'o', linewidth = '3')

plt.fill_between(ages_x, js_dev_y, alpha = 0.25 )
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by Age')
plt.legend() 
plt.show()
```
