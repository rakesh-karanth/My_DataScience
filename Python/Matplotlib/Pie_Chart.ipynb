```python
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
```

```python
plt.pie(slices)
plt.title('My pie chart')
plt.show()
```

```python
plt.pie(slices, labels =labels)
plt.title('My pie chart')
plt.show()
```

```python
plt.pie(slices, labels = labels, wedgeprops={'edgecolor':'black'})
plt.title('My pie chart')
plt.show()
```

### To add custom color

```python
slices = [59219, 55466, 47544, 36443]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python']
#color = ['blue', 'red', 'green', 'yellow']
# better use like below will be abvailable in matplotlib page.
color = ['#008fd5', '#fc4f30', '#e5ae37', '#6d904f']
```

```python
plt.pie(slices, colors = color, labels = labels, wedgeprops={'edgecolor':'black'})
plt.title('My pie chart')
plt.show()
```

# Exploding the particular piece

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]
```

```python
plt.pie(slices, colors = color, labels = labels, wedgeprops={'edgecolor':'black'}, explode = explode)
plt.title('My pie chart')
plt.show()
```

# To give some asthetic
- adding shadow 
- rotating the plot 

```python
plt.pie(
        slices, labels = labels, colors = color, 
        shadow=True, startangle=90,
        explode = explode, autopct='%1.1f%%',
        wedgeprops={'edgecolor':'black'}
       )
plt.title('My pie chart')
plt.show()
```
