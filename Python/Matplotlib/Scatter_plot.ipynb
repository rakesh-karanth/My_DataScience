Scatter plot usefull to find the corelation between two variables

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
x = [5,7,8,5,6,7,9,2,3,4,4,4,2,6,3,6,8,6,4,1]
y = [7,4,3,9,1,3,2,5,2,4,8,7,1,6,4,9,7,7,5,1]
```

# Basic scatter plot

```python
plt.scatter(x,y)
plt.title('Scatter plot')
plt.xlabel('x label')
plt.ylabel('y label')
plt.show()
```

# To change color and style

```python
plt.scatter(x,y, 
            s=100, c='green', 
            marker='*', 
            edgecolor = 'black',
            linewidth = 2, 
            alpha = 0.75)
```

# To give different colors to distinguish the charts

```python
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
```

```python
plt.scatter(x,y, 
            s=100, c=colors, cmap = 'Greens',
            edgecolor = 'black',
            linewidth = 1, 
            alpha = 0.75,
            )
```

# To add bar

```python
plt.scatter(x,y, 
            s=100, c=colors, cmap = 'Greens',
            edgecolor = 'black',
            linewidth = 1, 
            alpha = 0.75,
            )
cbar = plt.colorbar()
cbar.set_label('Satisfaction')
```

# Read the data from data set and plot using scatter plot to find the corelation between view count and total likes

```python
import pandas as pd
```

```python
data = pd.read_csv('2019-05-31-data.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, c=ratio, cmap='summer',
            edgecolor='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.xscale('log')
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()
```
