```python
# • Create new column <track_name_length> to store string length 
df['track_name_length'] = df['Track.Name'].apply(lambda x: len(x))
df.loc[:,['Track.Name','track_name_length']].head()
```

```python
# • Create new column <Artist_first_name> to store first name of each artist 
#Assumption: first substring is the first name
# 'The Chainsmokers'.split(' ')[0]
df['Artist_first_name'] = df['Artist.Name'].apply(lambda x: x.split(' ')[0])
df.loc[:,['Artist.Name','Artist_first_name']].head()
```

```python
# • How many songs from “dance pop” genre? 
print(df[df.Genre == 'dance pop'].shape[0])
print(len(df[df.Genre == 'dance pop']))
print(df[df.Genre == 'dance pop']['Track.Name'].count())

# if we wish to remove duplicates on Track.Name:
print(df[df.Genre == 'dance pop']['Track.Name'].nunique())
print(len(df[df.Genre == 'dance pop']['Track.Name'].unique()))
print(len(set(df[df.Genre == 'dance pop']['Track.Name'])))
```

```python
# • What is average length for each genre? 
df.groupby(by = 'Genre')[['Length.']].mean()
```

```python
# • What is average 'Length.' for each artist? 
df.groupby('Artist.Name')[['Length.']].mean()
```

```python
df = df.append({'Track.Name':'One Thing Right', 'Popularity':88,'Danceability':29}, ignore_index= True)
df.shape
```

```python
# • Rank each song based on popularity. 
df['popularity_rank'] = df.groupby('Popularity')['Track.Name'].transform(lambda x: x.rank())
df.loc[:,['Popularity','Track.Name','popularity_rank']]

## method = average (default value): takes count of records for each Popularity and assigns average value.
```

```python
# • Rank each song based on danceability. 
df['danceability_rank'] = df.groupby('Danceability')['Track.Name'].transform(lambda x: x.rank())
df.loc[:,['Danceability','Track.Name','danceability_rank']]
```

```python
# • Average, Maximum, Minimum, standard deviation of each column 
df.drop('Unnamed: 0',axis = 1).describe()
```

```python
numerical_columns = df.select_dtypes(include='number').columns
for operation in ['max','min','std']:
    print(df[numerical_columns].drop('Unnamed: 0',axis=1).apply(func = operation, axis = 0))
    print('\n')
```

```python
# • Average, Maximum, Minimum, standard deviation of each row 
numerical_columns = df.select_dtypes(include='number').columns
for operation in ['max','min','std']:
    print(f'operation is: {operation}')
    print(df[numerical_columns].drop('Unnamed: 0', axis = 1).apply(func = operation, axis = 1))
    print('\n')
```
